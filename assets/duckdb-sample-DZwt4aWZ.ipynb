# DuckDB in notebooks

Use the **DuckDB** kernel to run SQL in cells. Results are shown as text.

```
SELECT 'DuckDB' AS engine, version() AS v;
```

```
CREATE OR REPLACE TEMP TABLE t (id INT, name VARCHAR);
INSERT INTO t VALUES (1, 'a'), (2, 'b'), (3, 'c');
SELECT * FROM t;
```

```
SELECT name, id * 10 AS scaled FROM t ORDER BY id DESC;
```
