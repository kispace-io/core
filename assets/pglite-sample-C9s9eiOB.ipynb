# PostgreSQL (PGlite) in notebooks

Use the **PGlite** kernel to run PostgreSQL SQL in the browser.

```
SELECT 'PGlite' AS engine, version() AS v;
```

```
CREATE TABLE IF NOT EXISTS demo (id SERIAL PRIMARY KEY, label TEXT);
INSERT INTO demo (label) VALUES ('first'), ('second'), ('third');
SELECT * FROM demo;
```

```
SELECT label, id * 2 AS double_id FROM demo ORDER BY id;
```
